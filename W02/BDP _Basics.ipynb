## 7. Combining datasets

Sometimes, the datasets we find do not contain enough information for our project. Consequently, we need to merge datasets in order to construct a dataset which contains all the necessary information.

For instance, your 'video_games' dataset does not contain any information on the game genre. Below, we show you how to merge this dataset with a dataset containing game genres.

Frist, we import dataset **game_genres.csv** containing information about the video games' genres

```python
game_genres = pd.read_csv('genres.csv', delimiter=',')
game_genres
```

In the game_genres dataset there is also a column named 'Rank', we can use that column to merge with the 'Rank' column in the 'video_games' dataset (which we named 'df' when we imported it back in the beginning)

There are different ways of merging two dataframes, choosing "outer" is similar to a full outer join in SQL:
https://www.w3schools.com/sql/sql_join_full.asp

As you can see from the output of the code below, we have now created a dataframe that contains a column called 'Genre':

```python
vd_genres = pd.merge(df, game_genres, on=['Rank'], how='outer')
vd_genres
```

Another way to combine two datasets is by use of the function **.concat()**

```python
vd_data = [df, game_genres]
vd_genres2 = pd.concat(vd_data, sort=False, join='outer')
vd_genres2
```

```python
def three_country_comparison():
    #TIPS & TRICKS FOR VISUALISATION

    #Changing the size of the figure
    plt.figure(figsize=(10, 10))

    #Plotting a visualization .plot(x, y, data=data)
    plt.plot('year', 'lifeExp', data=denmark)
    plt.plot('year', 'lifeExp', data=mexico, marker = 'o')
    plt.plot('year', 'lifeExp', data=china, linestyle = 'dashed')

    #Giving the visualization a title
    plt.title("Three country comparison")

    #Giving the x-axis a name
    plt.xlabel("Year")

    #Giving the y-axis a name
    plt.ylabel("Life expectancy")

    #Adding a legend
    plt.legend(['Denmark', 'Mexico', 'China'])

    #Only a requirement outside of Jupyter Notebook
    plt.show()

three_country_comparison()
```
